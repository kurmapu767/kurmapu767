```
import numpy as np
import pandas as pd
from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from google.colab import files
upload= files.upload()
```

```
df=pd.read_csv("apple.txt",error_bad_lines=False)
df
```

```
for x in df.x:
  print(x)
```

```
df=[x.strip() for x in df.x]
df=[x for x in df if x]
```

```
text=' '.join(df)
```

```
text
```

```
#generate word cloud
from wordcloud import STOPWORDS
def plot_cloud(wordcloud):
  plt.figure(figsize=(15,30))
  plt.imshow(wordcloud)
  plt.axis("off")
stopwords=STOPWORDS
stopwords.add("will")
stopwords.add("apple")
stopwords.add("Amazon")
stopwords.add("laptop")
stopwords.add("got")
stopwords.add("product")
wordcloud=WordCloud(width=3000,height=2000,background_color='white',max_words=100,stopwords=stopwords).generate(text)
plot_cloud(wordcloud)
```
