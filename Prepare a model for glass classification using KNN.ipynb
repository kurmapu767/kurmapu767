```
df=pd.read_csv("glass.csv")
df.head(5)
```

```
y=df["Type"]
x=df.iloc[:,0:9]
list(x)
```

```
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
a= ss.fit_transform(x)
x=pd.DataFrame(a)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.34)




```

```
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_test_pred= knn.predict(x_test)
y_train_pred=knn.predict(x_train)
```

```
from sklearn.metrics import accuracy_score
asc=accuracy_score(y_train,y_train_pred)
bs=accuracy_score(y_test,y_test_pred)
print("train accuracy score:",asc)
print("test accuracy score:",bs)
```

```
train_acc=[]
test_acc=[]


for i in range(1,15):
  knn= KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  y_test_pred= knn.predict(x_test)
  y_train_pred=knn.predict(x_train)
  train_acc.append(accuracy_score(y_train,y_train_pred))
  test_acc.append(accuracy_score(y_test,y_test_pred))
d1=pd.DataFrame(range(1,15))
d2=pd.DataFrame(train_acc)
d3=pd.DataFrame(test_acc)
pd.concat([d1,d2,d3],axis=1)
```
